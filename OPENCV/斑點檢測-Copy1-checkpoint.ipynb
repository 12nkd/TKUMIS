```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#讀取影像
img_path = "photo2.jpg"
img = cv2.imread(img_path, cv2.IMREAD_COLOR) #IMREAD_COLOR是一個參數，表示要以彩色模式讀取圖片，即使圖片是灰階的也會轉為BGR

#初始化SimpleBlobDetector
params = cv2.SimpleBlobDetector_Params() #建立一個參數設定物件，用來定義你要偵測哪些樣子的blob，params可以儲存設定物件，接下來可以對它設定條件

#設定參數
params.filterByColor = False  
params.blobColor = 0  # 若設定0 就是檢測所有顏色

params.minThreshold = 100 #最小，二值化閥值
params.maxThreshold = 200 #最大，二值化閥值

params.filterByArea = True #只檢測設定的面積內
params.minArea = 40    #最小面積
params.maxArea = 1000   #最大面積

params.filterByCircularity = False #依照圓形程度過濾 blob
params.minCircularity = 0.5 #最小圓形度

#創建SimpleBlobDetector 
detector = cv2.SimpleBlobDetector_create(params)

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img2 = cv2.GaussianBlur(img2, (5, 5), 0)

keypoints = detector.detect(img2) #使用detector偵測img中的 blob，並把偵測到的結果儲存在keypoints裡
#逐一取出所有偵測到的blob，並印出它們的座標 (x, y) 和直徑大小
for point in keypoints:
    x, y = point.pt #(X,Y)座標
    x = round(x,2)
    y = round(y,2)
    Diameter = round(point.size,2) #直徑
    
    print(f"Keypoint at ({x}, {y}), Diameter: {Diameter}")

#  在圖片上畫出所有偵測到的斑點，並儲存在新圖片img_with_keypoints中
img_with_keypoints = cv2.drawKeypoints(img, keypoints, np.array([]), (0, 0, 255),cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 顯示原始圖像和帶有關鍵點的圖像
# cv2.imshow("Original Image", img)
# cv2.imshow("Image with Keypoints", img_with_keypoints)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# OpenCV 是 BGR，要轉成 RGB 才能正確顯示
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_kp_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_kp_rgb)
plt.title("Image with Keypoints")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#讀取影像
img_path = "photo2.jpg"
img = cv2.imread(img_path, cv2.IMREAD_COLOR) #IMREAD_COLOR是一個參數，表示要以彩色模式讀取圖片，即使圖片是灰階的也會轉為BGR

#初始化 SimpleBlobDetector
params = cv2.SimpleBlobDetector_Params() #建立一個參數設定物件，用來定義你要偵測哪些樣子的blob，params可以儲存設定物件，接下來可以對它設定條件

#設定參數
params.filterByColor = False  #只檢測指定的顏色
params.blobColor = 0  # 若設定0 就是檢測所有顏色

params.minThreshold = 10 #最小，二值化閥值
params.maxThreshold = 250 #最大，二值化閥值

params.filterByArea = True #只檢測設定的面積內
params.minArea = 40   #最小面積
params.maxArea = 1000   #最大面積

params.filterByCircularity = False #依照圓形程度過濾 blob
params.minCircularity = 0.5  #最小圓形度

#創建SimpleBlobDetector 
detector = cv2.SimpleBlobDetector_create(params)

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img2 = cv2.GaussianBlur(img2, (11, 11), 8.5)


keypoints = detector.detect(img2) #使用detector偵測img中的 blob，並把偵測到的結果儲存在keypoints裡
#逐一取出所有偵測到的blob，並印出它們的座標 (x, y) 和直徑大小
print(f"檢測到的關鍵點數量: {len(keypoints)}")  # 檢查檢測到的點數量
for point in keypoints:
    x, y = point.pt #(X,Y)座標
    x = round(x,2)
    y = round(y,2)
    Diameter = round(point.size,2) #直徑
    # 將座標轉為整數以獲取灰階值
    x_int, y_int = int(round(x)), int(round(y))
    # 確保座標在圖像範圍內
    if 0 <= y_int < img2.shape[0] and 0 <= x_int < img2.shape[1]:
        gray_value = img2[y_int, x_int]  # 獲取灰階值
    else:
        gray_value = None  # 如果座標超出範圍，設為 None

    print(f"Keypoint at ({x}, {y}), Diameter: {Diameter}, Gray Value: {gray_value}")
    #print(f"Keypoint at ({x}, {y}), Diameter: {Diameter}, ")

#  在圖片上畫出所有偵測到的斑點，並儲存在新圖片img_with_keypoints中
img_with_keypoints = cv2.drawKeypoints(img, keypoints, np.array([]), (0, 0, 255),cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 顯示原始圖像和帶有關鍵點的圖像
# cv2.imshow("Original Image", img)
# cv2.imshow("Image with Keypoints", img_with_keypoints)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# OpenCV 是 BGR，要轉成 RGB 才能正確顯示
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_kp_rgb = cv2.cvtColor(img_with_keypoints, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_kp_rgb)
plt.title("Image with Keypoints")
plt.axis('off')

plt.tight_layout()
plt.show()
```
