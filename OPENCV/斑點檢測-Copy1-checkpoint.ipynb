```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#
img_path = "photo2.jpg"
img = cv2.imread(img_path, cv2.IMREAD_COLOR) #IMREAD_COLORBGR

#SimpleBlobDetector
params = cv2.SimpleBlobDetector_Params() #blobparams

#
params.filterByColor = False  
params.blobColor = 0  # 0 

params.minThreshold = 100 #
params.maxThreshold = 200 #

params.filterByArea = True #
params.minArea = 40    #
params.maxArea = 1000   #

params.filterByCircularity = False # blob
params.minCircularity = 0.5 #

#SimpleBlobDetector 
detector = cv2.SimpleBlobDetector_create(params)

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img2 = cv2.GaussianBlur(img2, (5, 5), 0)

keypoints = detector.detect(img2) #detectorimg blobkeypoints
#blob (x, y) 
for point in keypoints:
    x, y = point.pt #(X,Y)
    x = round(x,2)
    y = round(y,2)
    Diameter = round(point.size,2) #
    
    print(f"Keypoint at ({x}, {y}), Diameter: {Diameter}")

#  img_with_keypoints
img_with_keypoints = cv2.drawKeypoints(img, keypoints, np.array([]), (0, 0, 255),cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 
# cv2.imshow("Original Image", img)
# cv2.imshow("Image with Keypoints", img_with_keypoints)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# OpenCV  BGR RGB 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_kp_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_kp_rgb)
plt.title("Image with Keypoints")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#
img_path = "photo2.jpg"
img = cv2.imread(img_path, cv2.IMREAD_COLOR) #IMREAD_COLORBGR

# SimpleBlobDetector
params = cv2.SimpleBlobDetector_Params() #blobparams

#
params.filterByColor = False  #
params.blobColor = 0  # 0 

params.minThreshold = 10 #
params.maxThreshold = 250 #

params.filterByArea = True #
params.minArea = 40   #
params.maxArea = 1000   #

params.filterByCircularity = False # blob
params.minCircularity = 0.5  #

#SimpleBlobDetector 
detector = cv2.SimpleBlobDetector_create(params)

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img2 = cv2.GaussianBlur(img2, (11, 11), 8.5)


keypoints = detector.detect(img2) #detectorimg blobkeypoints
#blob (x, y) 
print(f": {len(keypoints)}")  # 
for point in keypoints:
    x, y = point.pt #(X,Y)
    x = round(x,2)
    y = round(y,2)
    Diameter = round(point.size,2) #
    # 
    x_int, y_int = int(round(x)), int(round(y))
    # 
    if 0 <= y_int < img2.shape[0] and 0 <= x_int < img2.shape[1]:
        gray_value = img2[y_int, x_int]  # 
    else:
        gray_value = None  #  None

    print(f"Keypoint at ({x}, {y}), Diameter: {Diameter}, Gray Value: {gray_value}")
    #print(f"Keypoint at ({x}, {y}), Diameter: {Diameter}, ")

#  img_with_keypoints
img_with_keypoints = cv2.drawKeypoints(img, keypoints, np.array([]), (0, 0, 255),cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 
# cv2.imshow("Original Image", img)
# cv2.imshow("Image with Keypoints", img_with_keypoints)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# OpenCV  BGR RGB 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_kp_rgb = cv2.cvtColor(img_with_keypoints, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_kp_rgb)
plt.title("Image with Keypoints")
plt.axis('off')

plt.tight_layout()
plt.show()
```
